Task
Step1 : 

we first calculate the frequency of each genre in the Genre column using value_counts(). We then select the top 5 genres based on their frequency by using the head(5) function on the genre_counts Series.

Next, we filter the dataset vg based on the selected top 5 genres using the isin() function. This filters the data and keeps only the rows where the Genre column value is one of the top 5 genres.

```python
genre_counts = vg['Genre'].value_counts()
top5_genre = genre_counts.head(5).index
myData= vg[vg['Genre'].isin(top5_genre)]
myData 
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(data=myData, x='Genre', y='Global_Sales')
plt.title('Stripplot of Genre and Global Sales (Top 5 Genres)')
plt.show()
```

Create a barplot to display Publisher and the Global Sales. Data is filtered on the basis of the maximum frequent values inside the categorical cols. Select upto 5 values after filtration: 

```python
Publisher_counts = vg['Publisher'].value_counts()
top5_Publisher =Publisher_counts.head(5).index
myData= vg[vg['Publisher'].isin(top5_Publisher)]
myData 
```

```python
plt.figure(figsize=(12, 6))
sns.stripplot(data=myData, x='Publisher', y='Global_Sales')
plt.title('Stripplot of Publisher and Global Sales (Top 5 Genres)')
plt.show()
```

Create Functions that can create group on the basis of categorical columns and show the total collection of the sales.


```python
def calculate_total_sales(vg, Genre):
    grouped_sales = vg.groupby(Genre)['Global_Sales'].sum()
    return grouped_sales
```

```python
genre_sales = calculate_total_sales(myData, 'Genre')
publisher_sales = calculate_total_sales(myData, 'Publisher')

genre_sales
```

```python
publisher_sales
```

```python
publisher_sales = vg.groupby(['Publisher', 'Year'])[['NA_Sales', 'EU_Sales', 'JP_Sales']].sum().reset_index()
publisher_sales
```

Show the line chart for the gaming Publisher and display the overall collection of sales in NA, EU and JP with respect to the year in the same chart.

```python
plt.figure(figsize=(15, 10))
sns.lineplot(data=publisher_sales, x='Year', y='NA_Sales', label='NA Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Overall Sales by Region for Publishers')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.lineplot(data=publisher_sales, x='Year', y='EU_Sales', label='EU_Sales')

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Overall Sales by Region for Publishers')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.lineplot(data=publisher_sales, x='Year', y='JP_Sales', label='JP Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Overall Sales by Region for Publishers')
plt.legend()
plt.show()
```
